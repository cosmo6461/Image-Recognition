```python
import json
import pandas as pd
file = open("/content/drive/MyDrive/recognition.json")
```

```python
def flatten_dictionary(input_dict, parent_key = ''):
    '''
    This function turns
    '''
    flattened_dict = {}
    for k, v in input_dict.items():
      key = f"{parent_key}_{k}" if parent_key else k
      if isinstance(v, dict):
        flattened_dict.update(flatten_dictionary(v, key))
      elif isinstance(v, list):
        if v == []:
          flattened_dict[key] = 0
        for index, item in enumerate(v):
          if isinstance(item, dict):
            flattened_dict.update(flatten_dictionary(item, f"{key}_{index+1}"))
          else:
            flattened_dict[f"{key}_{index+1}"] = item
      else:
        flattened_dict[key] = v
    return flattened_dict
```

```python
def construct_data_frame(file):
  data = json.load(file)
  data = data['Labels']
  lst = [flatten_dictionary(dictionary) for dictionary in data]
  df = pd.DataFrame(lst)
  return df
print(construct_data_frame(file))
```
